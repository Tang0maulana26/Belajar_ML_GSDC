```python
import numpy as np #perhitungan
df[['X1','X10','X20','X50','X70','X90','X100','X120','X128']] = df[['X1','X10','X20','X50','X70','X90','X100','X120','X128']].replace(0,np.NaN)
```

Data PREPROCESSING

```python
#Preprocessing
#handling missing value
df['X1'].fillna(df['X1'].mean(), inplace = True)
df['X10'].fillna(df['X10'].mean(), inplace = True)
df['X20'].fillna(df['X20'].mean(), inplace = True)
df['X50'].fillna(df['X50'].mean(), inplace = True)
df['X70'].fillna(df['X70'].mean(), inplace = True)
df['X90'].fillna(df['X90'].mean(), inplace = True)
df['X100'].fillna(df['X100'].mean(), inplace = True)
df['X120'].fillna(df['X120'].mean(), inplace = True)
df['X128'].fillna(df['X128'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_rows = (len(df.columns)+2) // 3
fig, axes = plt.subplots(nrows = num_rows, ncols = 3, figsize=(20,35))

for i, col in enumerate(df.columns):
  sns.boxplot(data=df, x=col, ax=axes[i // 3 , i % 3])

plt.tight_layout()
plt.show()
```

```python
#Removing ouliers
outliers = df.quantile(.97) # dealing with the outliers seen in the boxplot above

df = df[(df['X1']<outliers['X1'])]
df = df[(df['X10']<outliers['X10'])]
df = df[(df['X20']<outliers['X20'])]
df = df[(df['X50']<outliers['X50'])]
df = df[(df['X70']<outliers['X70'])]
df = df[(df['X90']<outliers['X90'])]
df = df[(df['X100']<outliers['X100'])]
df = df[(df['X120']<outliers['X120'])]
df = df[(df['X128']<outliers['X128'])]
```

```python
#Data Allocating
y = df['Unnamed: 0']
x = df.drop(['Unnamed: 0'], axis=1)
```

```python
#Standart Scalar -> menyelesaikan fitur" data sehinggan pada range yang tepat
from sklearn.preprocessing import StandardScaler

col = x.columns
scaler = StandardScaler()

x = scaler.fit_transform(x)
x = pd.DataFrame(data = x, columns = col)

x
```

Feature Engineering

```python
from sklearn.decomposition import PCA

pca = PCA(n_components= 0.95)
x = pca.fit_transform(x)

x
```

```python
#Spliting dataset
from sklearn.model_selection import train_test_split

#Pisahin dataset Train: 75%, test : 25%
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
print(y_test.value_counts())
```

```python
#Balancing Dataset
from imblearn.over_sampling import RandomOverSampler

over = RandomOverSampler()
x, y = over.fit_resample(x_train, y_train)

print(y.value_counts())
 
```

Modelling

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(x_train, y_train)
```

```python
from sklearn import metrics

predictionsrfc = rfc.predict(x_test)
print("Accuracy_score =", format(metrics.accuracy_score(y_test, predictionsrfc)))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, predictionsrfc))
print(classification_report(y_test, predictionsrfc))
```
